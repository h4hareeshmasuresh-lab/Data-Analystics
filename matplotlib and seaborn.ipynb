```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
read=pd.read_csv('book2.csv')
print(read)
line_plot=sns.lineplot(x="months",y="sales",data=read)
line_plot.set_title("company")
line_plot.set_xlabel("months")
line_plot.set_ylabel("sales")
plt.grid(True)
plt.show()
```
