```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
read=pd.read_csv('book1.csv')
print(read)
line_plot=sns.lineplot(x="name",y="age",data=read)
line_plot.set_title("age by name")
line_plot.set_xlabel("name")
line_plot.set_ylabel("age")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
read=pd.read_csv('book1.csv')
print(read)
scatter_plot=sns.scatterplot(x="age",y="salary",hue="name",data=read)
scatter_plot.set_title("age by salary")
scatter_plot.set_xlabel("name")
scatter_plot.set_ylabel("age")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
read=pd.read_csv('book1.csv')
print(read)
bar_chart=sns.barplot(x="name",y="age",data=read,palette="muted")
bar_chart.set_title("age by name")
bar_chart.set_xlabel("name")
bar_chart.set_ylabel("age")
plt.show()
```
